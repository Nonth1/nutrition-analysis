```python
# ============================================
# Nutrition Insight (Fixed: better column detection + numeric cleaning)
# ============================================
import os, re, io, math, json, datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

try:
    from google.colab import files
except Exception:
    files = None

pd.set_option('display.max_rows', 50)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 120)

# ---------- Load CSV ----------
def safe_read_csv(f, **kw):
    encodings = [None, 'utf-8', 'utf-8-sig', 'cp874', 'cp1252', 'latin1']
    last_err = None
    for enc in encodings:
        try:
            return pd.read_csv(f, encoding=enc, **kw)
        except Exception as e:
            last_err = e
    raise last_err

DF = None
if files is not None:
    print("üì• ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡πÑ‡∏ü‡∏•‡πå CSV ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏≠‡∏±‡∏õ‡πÇ‡∏´‡∏•‡∏î (‡∏´‡∏£‡∏∑‡∏≠‡∏Ç‡πâ‡∏≤‡∏°‡∏ñ‡πâ‡∏≤‡∏à‡∏∞‡πÉ‡∏ä‡πâ path)...")
    uploaded = files.upload()
    if uploaded:
        fname = list(uploaded.keys())[0]
        print(f"Loaded: {fname}")
        DF = safe_read_csv(io.BytesIO(uploaded[fname]))

if DF is None:
    default_path = "/content/nutrition.csv"
    if os.path.exists(default_path):
        DF = safe_read_csv(default_path)
        print(f"Loaded from path: {default_path}")
    else:
        raise FileNotFoundError("‡πÑ‡∏°‡πà‡∏û‡∏ö‡πÑ‡∏ü‡∏•‡πå CSV (‡∏≠‡∏±‡∏õ‡πÇ‡∏´‡∏•‡∏î‡πÑ‡∏ü‡∏•‡πå‡∏´‡∏£‡∏∑‡∏≠‡πÅ‡∏Å‡πâ default_path)")

print("Rows:", len(DF), "| Cols:", len(DF.columns))
display(DF.head(3))

# ---------- Helpers: column detection & numeric cleaning ----------
def norm_name(s: str) -> str:
    s = str(s).lower()
    s = re.sub(r"[()\[\]{}:_\-\/\\,.;%]+", " ", s)
    s = re.sub(r"\s+", " ", s).strip()
    return s

def clean_numeric_series(s: pd.Series):
    if s is None: return None
    s = s.astype(str)
    s = s.str.replace(",", "", regex=False)           # "1,234" -> "1234"
    s = s.str.replace("\u00a0", " ", regex=False)     # NBSP
    s = s.str.extract(r"([-+]?\d*\.?\d+)")[0]         # "3.5 g" -> "3.5"
    return pd.to_numeric(s, errors="coerce")

ALIASES = {
    "name":  ["name","food","product","menu","item","description","short description","long description"],
    "cal":   ["calorie","calories","kcal","energy","energy kcal"],
    "protein":["protein","proteins","prot"],
    "fat":   ["fat","total fat","lipid","total lipid"],
    "sugar": ["sugar","sugars","total sugar","total sugars"],
    "fiber": ["fiber","fibre","dietary fiber","dietary fibre"],
    "carb":  ["carb","carbs","carbohydrate","carbohydrates","cho","carbohydrate by difference"],
    "serv_g":["serving size g","serv size g","serving g","weight g","per serving g","serving weight g","serving_size_g"]
}
# ‡∏ö‡∏≤‡∏á‡∏ä‡∏∑‡πà‡∏≠‡∏Ñ‡∏ß‡∏£‡∏´‡∏•‡∏µ‡∏Å‡πÄ‡∏•‡∏µ‡πà‡∏¢‡∏á‡πÄ‡∏ß‡∏•‡∏≤‡∏à‡∏±‡∏ö calories ‡πÄ‡∏ä‡πà‡∏ô "calories from fat"
NEGATIVE = {
    "cal": ["from fat"]
}

def find_best_numeric_column(df: pd.DataFrame, key: str, default=None):
    """‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ó‡∏µ‡πà '‡∏ä‡∏∑‡πà‡∏≠‡∏™‡∏∑‡πà‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏´‡∏°‡∏≤‡∏¢' + '‡∏°‡∏µ‡∏Ñ‡πà‡∏≤ numeric non-null ‡∏°‡∏≤‡∏Å‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î'"""
    cand_tokens = [norm_name(a) for a in ALIASES.get(key, [])]
    neg_tokens  = [norm_name(a) for a in NEGATIVE.get(key, [])] if key in NEGATIVE else []
    scored = []

    for c in df.columns:
        nc = norm_name(c)
        if any(tok in nc for tok in cand_tokens) and not any(bad in nc for bad in neg_tokens):
            ser_clean = clean_numeric_series(df[c]) if key != "name" else df[c].astype(str)
            non_null = ser_clean.notna().sum()
            scored.append((non_null, c))

    if scored:
        scored.sort(reverse=True)          # ‡πÄ‡∏≠‡∏≤ non-null ‡∏°‡∏≤‡∏Å‡∏™‡∏∏‡∏î
        best = scored[0][1]
        return best

    # ‡∏ñ‡πâ‡∏≤‡πÑ‡∏°‡πà‡πÄ‡∏à‡∏≠‡πÄ‡∏•‡∏¢‡πÅ‡∏ï‡πà‡∏≠‡∏¢‡∏≤‡∏Å fallback
    return default

# ---------- Map columns ----------
col_name = find_best_numeric_column(DF, "name", default=DF.columns[0])
col_cal  = find_best_numeric_column(DF, "cal")
col_pro  = find_best_numeric_column(DF, "protein")
col_fat  = find_best_numeric_column(DF, "fat")
col_sug  = find_best_numeric_column(DF, "sugar")
col_fib  = find_best_numeric_column(DF, "fiber")
col_carb = find_best_numeric_column(DF, "carb")
col_serv = find_best_numeric_column(DF, "serv_g")

found = {
    "name": col_name, "cal": col_cal, "protein": col_pro,
    "fat": col_fat, "sugar": col_sug, "fiber": col_fib,
    "carb": col_carb, "serving_g": col_serv
}
print("Detected columns:", json.dumps(found, ensure_ascii=False, indent=2))

# ---------- Keep columns & clean numerics ----------
use_cols = [c for c in [col_name, col_cal, col_pro, col_fat, col_sug, col_fib, col_carb, col_serv] if c is not None]
DF = DF[use_cols].copy()
DF.rename(columns={col_name: "Food"}, inplace=True)

for c in [col_cal, col_pro, col_fat, col_sug, col_fib, col_carb, col_serv]:
    if c and c in DF.columns:
        if c != "Food":
            DF[c] = clean_numeric_series(DF[c])

# ‡∏™‡∏£‡∏∏‡∏õ‡∏Ñ‡∏∏‡∏ì‡∏†‡∏≤‡∏û‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏´‡∏•‡∏±‡∏á‡∏ó‡∏≥‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏≠‡∏≤‡∏î
diag = {}
for k, c in [("cal",col_cal),("protein",col_pro),("fat",col_fat),("sugar",col_sug),("fiber",col_fib),("carb",col_carb)]:
    if c and c in DF.columns:
        diag[k] = int(DF[c].notna().sum())
print("Non-null counts after cleaning:", diag)

display(DF.head(5))

# ---------- Normalize to per 100g ----------
def typical_already_per100g(cal_series):
    s = cal_series.dropna()
    if len(s) == 0: return False
    med = s.median()
    return 10 <= med <= 600

X = DF.copy()
if col_cal is None:
    raise ValueError("‡πÑ‡∏°‡πà‡∏û‡∏ö‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå Calories/Energy")

already_per100 = typical_already_per100g(X[col_cal])

if col_serv and col_serv in X.columns and not already_per100:
    grams = X[col_serv].where(X[col_serv] > 0)
    factor = 100.0 / grams
    for c in [col_cal, col_pro, col_fat, col_sug, col_fib, col_carb]:
        if c and c in X.columns:
            X[c + "_per100g"] = X[c] * factor
else:
    for c in [col_cal, col_pro, col_fat, col_sug, col_fib, col_carb]:
        if c and c in X.columns:
            X[c + "_per100g"] = X[c]

X["cal_per_g"] = X[col_cal + "_per100g"] / 100.0
keep_cols = ["Food"] + [c for c in X.columns if c.endswith("_per100g")] + ["cal_per_g"]
display(X.head(8)[keep_cols])

# ---------- Thresholds ----------
def q(series, p):
    s = series.dropna()
    return s.quantile(p) if len(s) else np.nan

CAL = X[col_cal + "_per100g"]
SUG = X[col_sug + "_per100g"] if col_sug else pd.Series(index=X.index, dtype=float)
PRO = X[col_pro + "_per100g"] if col_pro else pd.Series(index=X.index, dtype=float)
FAT = X[col_fat + "_per100g"] if col_fat else pd.Series(index=X.index, dtype=float)
FIB = X[col_fib + "_per100g"] if col_fib else pd.Series(index=X.index, dtype=float)

FIX = {
    "high_cal_kcal_per100": 250,
    "high_sugar_g_per100":  15,
    "high_protein_g_per100":15,
    "low_fat_g_per100":     5,
    "high_fiber_g_per100":  5,
    "low_cal_kcal_per100":  100
}
Q = {
    "cal_high": q(CAL, 0.70),
    "sug_high": q(SUG, 0.70) if col_sug else np.nan,
    "pro_high": q(PRO, 0.70) if col_pro else np.nan,
    "fat_low":  q(FAT, 0.40) if col_fat else np.nan,
    "fib_high": q(FIB, 0.70) if col_fib else np.nan,
    "cal_low":  q(CAL, 0.30)
}
def T_fix_or_q(fixed, qval, prefer_fixed=True):
    if prefer_fixed and not math.isnan(fixed): return fixed
    return qval if not math.isnan(qval) else fixed
TH = {
    "HIGH_CAL": T_fix_or_q(FIX["high_cal_kcal_per100"], Q["cal_high"]),
    "HIGH_SUG": T_fix_or_q(FIX["high_sugar_g_per100"], Q["sug_high"]),
    "HIGH_PRO": T_fix_or_q(FIX["high_protein_g_per100"], Q["pro_high"]),
    "LOW_FAT":  T_fix_or_q(FIX["low_fat_g_per100"], Q["fat_low"]),
    "HIGH_FIB": T_fix_or_q(FIX["high_fiber_g_per100"], Q["fib_high"]),
    "LOW_CAL":  T_fix_or_q(FIX["low_cal_kcal_per100"], Q["cal_low"]),
}
print("Heuristic thresholds (per 100g):", json.dumps(TH, indent=2))

# ---------- Segments ----------
def top_n(df, n=30): return df.head(n).copy()
out = {}

# A) Avoid: High Cal & High Sugar
if col_sug:
    mask_avoid = (X[col_cal + "_per100g"] >= TH["HIGH_CAL"]) & (X[col_sug + "_per100g"] >= TH["HIGH_SUG"])
    avoid_df = X.loc[mask_avoid, ["Food", col_cal + "_per100g", col_sug + "_per100g"]].copy()
    avoid_df.rename(columns={col_cal + "_per100g":"Calories_100g", col_sug + "_per100g":"Sugar_g_100g"}, inplace=True)
    avoid_df = avoid_df.sort_values(["Calories_100g","Sugar_g_100g"], ascending=[False, False])
    out["avoid_high_cal_sugar"] = top_n(avoid_df, 50)
else:
    out["avoid_high_cal_sugar"] = pd.DataFrame(columns=["Food","Calories_100g","Sugar_g_100g"])
    print("‚ö†Ô∏è ‡πÑ‡∏°‡πà‡∏û‡∏ö‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ô‡πâ‡∏≥‡∏ï‡∏≤‡∏•")

# B) High Protein, Low Fat
if col_pro and col_fat:
    mask_hplf = (X[col_pro + "_per100g"] >= TH["HIGH_PRO"]) & (X[col_fat + "_per100g"] <= TH["LOW_FAT"])
    hplf_df = X.loc[mask_hplf, ["Food", col_pro + "_per100g", col_fat + "_per100g", col_cal + "_per100g"]].copy()
    hplf_df.rename(columns={
        col_pro + "_per100g":"Protein_g_100g",
        col_fat + "_per100g":"Fat_g_100g",
        col_cal + "_per100g":"Calories_100g"
    }, inplace=True)
    hplf_df["Protein_per_Calorie"] = hplf_df["Protein_g_100g"] / (hplf_df["Calories_100g"].replace(0,np.nan))
    hplf_df = hplf_df.sort_values(["Protein_g_100g","Fat_g_100g","Calories_100g"], ascending=[False, True, True])
    out["high_protein_low_fat"] = top_n(hplf_df, 50)
else:
    out["high_protein_low_fat"] = pd.DataFrame(columns=["Food","Protein_g_100g","Fat_g_100g","Calories_100g","Protein_per_Calorie"])
    print("‚ö†Ô∏è ‡πÑ‡∏°‡πà‡∏û‡∏ö‡πÇ‡∏õ‡∏£‡∏ï‡∏µ‡∏ô‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏Ç‡∏°‡∏±‡∏ô")

# C) High Fiber, Low Calories
if col_fib:
    mask_hflc = (X[col_fib + "_per100g"] >= TH["HIGH_FIB"]) & (X[col_cal + "_per100g"] <= TH["LOW_CAL"])
    hflc_df = X.loc[mask_hflc, ["Food", col_fib + "_per100g", col_cal + "_per100g"]].copy()
    hflc_df.rename(columns={col_fib + "_per100g":"Fiber_g_100g", col_cal + "_per100g":"Calories_100g"}, inplace=True)
    hflc_df = hflc_df.sort_values(["Fiber_g_100g","Calories_100g"], ascending=[False, True])
    out["high_fiber_low_cal"] = top_n(hflc_df, 50)
else:
    out["high_fiber_low_cal"] = pd.DataFrame(columns=["Food","Fiber_g_100g","Calories_100g"])
    print("‚ö†Ô∏è ‡πÑ‡∏°‡πà‡∏û‡∏ö‡πÑ‡∏ü‡πÄ‡∏ö‡∏≠‡∏£‡πå")

for k,v in out.items():
    print("\n===", k, "===")
    display(v.head(10))

# ---------- (Optional) Plots ----------
if col_sug:
    plt.figure()
    plt.scatter(X[col_cal + "_per100g"], X[col_sug + "_per100g"], alpha=0.6)
    if not out["avoid_high_cal_sugar"].empty:
        a = out["avoid_high_cal_sugar"]
        plt.scatter(a["Calories_100g"], a["Sugar_g_100g"], alpha=0.9, marker='x')
    plt.axvline(TH["HIGH_CAL"], linestyle='--'); plt.axhline(TH["HIGH_SUG"], linestyle='--')
    plt.xlabel("Calories per 100g"); plt.ylabel("Sugar (g) per 100g")
    plt.title("High Calories & High Sugar (Avoid)"); plt.show()

if col_pro and col_fat:
    plt.figure()
    plt.scatter(X[col_pro + "_per100g"], X[col_fat + "_per100g"], alpha=0.6)
    if not out["high_protein_low_fat"].empty:
        h = out["high_protein_low_fat"]
        plt.scatter(h["Protein_g_100g"], h["Fat_g_100g"], alpha=0.9, marker='x')
    plt.axvline(TH["HIGH_PRO"], linestyle='--'); plt.axhline(TH["LOW_FAT"], linestyle='--')
    plt.xlabel("Protein (g) per 100g"); plt.ylabel("Fat (g) per 100g")
    plt.title("High Protein, Low Fat"); plt.show()

if col_fib:
    plt.figure()
    plt.scatter(X[col_fib + "_per100g"], X[col_cal + "_per100g"], alpha=0.6)
    if not out["high_fiber_low_cal"].empty:
        f = out["high_fiber_low_cal"]
        plt.scatter(f["Fiber_g_100g"], f["Calories_100g"], alpha=0.9, marker='x')
    plt.axvline(TH["HIGH_FIB"], linestyle='--'); plt.axhline(TH["LOW_CAL"], linestyle='--')
    plt.xlabel("Fiber (g) per 100g"); plt.ylabel("Calories per 100g")
    plt.title("High Fiber, Low Calories"); plt.show()

# ---------- Save CSV ----------
ts = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
paths = {}
for key, df in {
    "avoid_high_cal_sugar": out["avoid_high_cal_sugar"],
    "high_protein_low_fat": out["high_protein_low_fat"],
    "high_fiber_low_cal":   out["high_fiber_low_cal"]
}.items():
    p = f"{key}_{ts}.csv"
    df.to_csv(p, index=False)
    paths[key] = p
print("Saved files:", json.dumps(paths, indent=2))
```
